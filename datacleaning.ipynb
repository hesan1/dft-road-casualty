```python
import pandas as pd
```

```python
df = pd.read_csv("dft-road-casualty-statistics-casualty-provisional-mid-year-unvalidated-2022 (1).csv")
```

```python
df
```

```python
df = df.drop(['status', 'accident_year'], axis=1)
```

```python
df
```

```python
for column in df.columns:
    df = df[df[column] != -1]
```

```python
df
```

```python
# Check for null values
null_values = df.isnull().sum()
print("Null Values:\n", null_values)

# Check for duplicates
duplicates = df.duplicated()
print("\nNumber of Duplicates:", duplicates.sum())
```

```python
# Display unique values for each column
for column in df.columns:
    unique_values = df[column].unique()
    print(f"\nUnique values in {column}:\n", unique_values)
```

```python
df
```

```python
# Assuming your DataFrame is named 'df'
df = df[~df['vehicle_reference'].isin([227, 61])]
df = df[~df['casualty_reference'].isin([22, 148])]
```

```python
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Display distribution and relationships for Casualty_Class, Sex_of_Casualty, Age_of_Casualty, Age_Band_of_Casualty, Casualty_Severity
selected_columns = ['casualty_class', 'sex_of_casualty', 'age_of_casualty', 'age_band_of_casualty', 'casualty_severity']
for column in selected_columns:
    print(f"\n{column} - Value Counts:\n", df[column].value_counts())
    sns.countplot(x=column, data=df)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.show()
```

```python
# Display distribution for Pedestrian_Location, Pedestrian_Movement, Car_Passenger, Bus_or_Coach_Passenger, Pedestrian_Road_Maintenance_Worker
selected_columns = ['pedestrian_location', 'pedestrian_movement', 'car_passenger', 'bus_or_coach_passenger', 'pedestrian_road_maintenance_worker']
for column in selected_columns:
    print(f"\n{column} - Value Counts:\n", df[column].value_counts())
    sns.countplot(x=column, data=df)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.show()
```

```python
# Display distribution for Casualty_Type, Casualty_Home_Area_Type
selected_columns = ['casualty_type', 'casualty_home_area_type']
for column in selected_columns:
    print(f"\n{column} - Value Counts:\n", df[column].value_counts())
    sns.countplot(x=column, data=df)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.show()
```

```python
# Select numerical columns for correlation analysis
numerical_columns = ['age_of_casualty', 'casualty_severity']

# Calculate correlation matrix
correlation_matrix = df[numerical_columns].corr()

# Visualize the correlation matrix using a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# Select all numerical columns for correlation analysis
all_numerical_columns = df.select_dtypes(include='number').columns

# Calculate correlation matrix
correlation_matrix_all = df[all_numerical_columns].corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix_all, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```
